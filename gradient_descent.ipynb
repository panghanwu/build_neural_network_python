$w' = w - \eta\dfrac{\partial Loss}{\partial w}$  
$w\;:\; weight\\
w':\; updated\; weight\\
\eta\:\;:\; learning\; rate
$

# One Dimension

```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 2, 0.1)

def y(x):
  return (x-1)**2

plt.plot(x, y(x))
```

```
# differential y
def derived_y(x):
  return 2 * (x-1)

# parameter
x_p = 0.1  # initial position
eta = 0.1  # learning rate

# plot line
plt.plot(x, y(x))
plt.scatter(x_p, y(x_p))

# run gradient descent
while abs(derived_y(x_p)) > 0.1:
  x_p -= eta * derived_y(x_p)
  plt.scatter(x_p, y(x_p))
```

# Two Dimension

```
fig, ax = plt.subplots(figsize=(10,8))

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
xx, yy = np.meshgrid(x, y, sparse=True)

z = (xx-2)**2 + 0.2*(yy+2)**2

c = plt.contourf(x, y, z, 30)

ax.set_aspect('equal')
fig.colorbar(c, ax=ax)
```

```
def derived_z(x, y):
  return 2*(x-2), 1*(y+2)

# parameter
x_p = -4.  # initial position
y_p = 4.
eta = 0.6  # learning rate

# plot line
fig, ax = plt.subplots(figsize=(10,8))
c = plt.contourf(x, y, z, 30)
ax.set_aspect('equal')
fig.colorbar(c, ax=ax)
s = plt.scatter(x_p, y_p)


# run gradient descent
loss = 10.
while loss > .1:
  x_p_ = x_p
  y_p_ = y_p
  loss = abs(derived_z(x_p, y_p)[0]) + abs(derived_z(x_p, y_p)[1])
  # update
  x_p -= eta * derived_z(x_p, y_p)[0]
  y_p -= eta * derived_z(x_p, y_p)[1] 
  # plot
  plt.plot([x_p_, x_p], [y_p_, y_p], 'g-')
  plt.scatter(x_p, y_p)
```

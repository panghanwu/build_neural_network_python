$
\begin{bmatrix}
y_0\\
y_1
\end{bmatrix} = 
\begin{bmatrix}
x_0 & x_1
\end{bmatrix}
\begin{bmatrix}
w_{00} & w_{01}\\
w_{10} & w_{11}
\end{bmatrix}
$

```
import numpy as np
import matplotlib.pyplot as plt

# 100 input data
x_0 = np.arange(-1.0, 1.0, 0.2)
x_1 = np.arange(-1.0, 1.0, 0.2)
# container of 100 output 
z = np.zeros(100)


# define function of layers
def middle_layer(x, w, b):
  u = np.dot(x, w) + b
  return 1/(1+np.exp(-u))

def output_layer(x, w, b):
  u = np.dot(x, w) + b
  return u 
```

```
"""combination 1"""
# weights
w_im = np.array([[6.0,6.0], 
          [6.0,6.0]])  # input to middle
w_mo = np.array([[1.0],
          [-1.0]])  # middle to output
# biases
b_im = np.array([3.0,-3.0])
b_mo = np.array([0.1])


for i in range(10):
  for j in range(10):
    # generate input data
    inp = np.array([x_0[i], x_1[j]])
    # middle layer
    mid = middle_layer(inp, w_im, b_im)
    # output layer
    out = output_layer(mid, w_mo, b_mo)
    # outcome
    z[i*10+j] = out[0]  # turn array to value

plt.imshow(z.reshape(10,10), 'gray', vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```

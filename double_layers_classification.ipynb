```
import numpy as np
import matplotlib.pyplot as plt

# 400 input data
x_0 = np.arange(-1.0, 1.0, 0.1)
x_1 = np.arange(-1.0, 1.0, 0.1)
# container of 400*2 output 
z = np.zeros((400, 2))


# define function of layers
def middle_layer(x, w, b):
  u = np.dot(x, w) + b
  # sigmoid
  return 1/(1+np.exp(-u))

def output_layer(x, w, b):
  u = np.dot(x, w) + b
  # softmax
  return np.exp(u)/np.sum(np.exp(u))
```

```
"""combination 1"""
# weights
w_im = np.array([[1.0,2.0], 
          [3.0,3.0]])  # input to middle
w_mo = np.array([[-1.0,1.0],
          [1.0,-1.0]])  # middle to output
# biases
b_im = np.array([0.3,-0.3])
b_mo = np.array([0.4,0.1])

show_n = 5
for i in range(20):
  for j in range(20):
    # generate input data
    inp = np.array([x_0[i], x_1[j]])
    # middle layer
    mid = middle_layer(inp, w_im, b_im)
    # output layer
    out = output_layer(mid, w_mo, b_mo)
    # outcome
    z[i*20+j] = out
    if i*20+j < show_n:
      print(str(i*20+j) + ': (%.6f, %.6f)' % (z[i*20 + j][0], z[i*20 + j][1]))

# container of markers
triangle_x_0 = []
triangle_x_1 = []
circle_x_0 = []
circle_x_1 = []

# classify
for i in range(20):
  for j in range(20):
    if z[i*20+j][0] > z[i*20+j][1]:
      triangle_x_0.append(x_0[i])
      triangle_x_1.append(x_1[j])
    else:
      circle_x_0.append(x_0[i])
      circle_x_1.append(x_1[j])

# plot scatter
plt.scatter(triangle_x_0, triangle_x_1, marker='^')
plt.scatter(circle_x_0, circle_x_1, marker='o')
plt.show()
```
